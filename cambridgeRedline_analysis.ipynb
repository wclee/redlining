```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, mannwhitneyu
import numpy as np
```

```python
# Load the CSV file
file_path = "data/RedliningAccountability_3-25-25edit.csv"
df = pd.read_csv(file_path)
```

```python
# Clean and rename relevant columns
df_clean = df.rename(columns={
    'Unnamed: 0': 'School',
    'Unnamed: 1': 'RedlineStatus'
})

# Keep only relevant columns
df_clean = df_clean[['School', 'RedlineStatus', '2024', '2023', '2022', '2019', '2018']]

# Drop rows with missing status
df_clean = df_clean.dropna(subset=['RedlineStatus'])

# Melt the dataframe to long format
df_melted = df_clean.melt(id_vars=['School', 'RedlineStatus'], 
                          value_vars=['2024', '2023', '2022', '2019', '2018'],
                          var_name='Year', value_name='Score')

# Drop missing scores
df_melted = df_melted.dropna(subset=['Score'])

# Ensure numeric scores
df_melted['Score'] = pd.to_numeric(df_melted['Score'])
```

```python
# Independent t-test per year
ttest_results = {}
for year in df_melted['Year'].unique():
    redline_scores = df_melted[(df_melted['RedlineStatus'] == 'redline') & (df_melted['Year'] == year)]['Score']
    non_redline_scores = df_melted[(df_melted['RedlineStatus'] != 'redline') & (df_melted['Year'] == year)]['Score']
    if not redline_scores.empty and not non_redline_scores.empty:
        t_stat, p_val = ttest_ind(redline_scores, non_redline_scores, equal_var=False)
        ttest_results[year] = {'t_stat': t_stat, 'p_value': p_val}

# Print results
print("T-Test Results (Redline vs Non-Redline):")
for year, result in ttest_results.items():
    print(f"{year}: t={result['t_stat']:.2f}, p-value={result['p_value']:.3f}")
```

```python
# Custom color palette
custom_palette = {'redline': 'red', 'not redline': 'lightblue'}

# Create a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x='Year', y='Score', hue='RedlineStatus', palette=custom_palette)

plt.title("Score Comparison: Redline vs Non-Redline Schools")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.tight_layout()

# Move legend outside the plot
plt.legend(title="Redline Status", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Custom color palette
custom_palette = {'redline': 'red', 'not redline': 'lightblue'}

# Define consistent order for x-axis (years)
year_order = ['2018', '2019', '2022', '2023', '2024']

plt.figure(figsize=(12, 6))

# Draw boxplot with defined order
ax = sns.boxplot(
    data=df_melted,
    x='Year',
    y='Score',
    hue='RedlineStatus',
    palette=custom_palette,
    order=year_order,
    zorder=1  # Draw under points
)

# Plot individual school dots using the same order
for i, row in df_melted.iterrows():
    x_pos = year_order.index(row['Year'])  # Use defined order, not sorted()
    shift = -0.2 if row['RedlineStatus'] == 'redline' else 0.2
    x_jittered = x_pos + shift

    plt.scatter(
        x_jittered, row['Score'],
        color='black',
        s=30,
        alpha=0.9,
        edgecolor='black',
        zorder=2
    )

    plt.text(
        x_jittered, row['Score'] + 1,
        row['School'],
        ha='center',
        fontsize=7,
        rotation=45,
        zorder=3
    )

# Adjust axis ticks to match year_order
ax.set_xticks(range(len(year_order)))
ax.set_xticklabels(year_order)

# Adjust aesthetics
plt.title("Score Comparison: Redline vs Non-Redline Schools with Individual Labels")
plt.ylabel("Score")

# Move legend outside
handles, labels = ax.get_legend_handles_labels()
ax.legend(
    handles[:2], labels[:2],  # Remove duplicate legends
    title="Redline Status",
    bbox_to_anchor=(1.02, 1),
    loc='upper left',
    borderaxespad=0
)

plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

```python
# Load the CSV file
file_path = "data/RedliningAccountability_3-25-25edit.csv"  # Adjust path as needed
df = pd.read_csv(file_path)

df_clean = df.rename(columns={'Unnamed: 0': 'School', 'Unnamed: 1': 'RedlineStatus'})
df_clean = df_clean[['School', 'RedlineStatus', '2024', '2023', '2022', '2019', '2018']]
df_clean = df_clean.dropna(subset=['RedlineStatus'])

df_melted = df_clean.melt(
    id_vars=['School', 'RedlineStatus'],
    value_vars=['2024', '2023', '2022', '2019', '2018'],
    var_name='Year',
    value_name='Score'
)

df_melted = df_melted.dropna(subset=['Score'])
df_melted['Score'] = pd.to_numeric(df_melted['Score'])

# Group data
grouped = df_melted[['RedlineStatus', 'Score']]
redline_scores = grouped[grouped['RedlineStatus'] == 'redline']['Score']
non_redline_scores = grouped[grouped['RedlineStatus'] != 'redline']['Score']

# Stats
t_stat, t_pval = ttest_ind(redline_scores, non_redline_scores, equal_var=False)
u_stat, u_pval = mannwhitneyu(redline_scores, non_redline_scores, alternative='two-sided')

# Plot
plt.figure(figsize=(8, 6))

# Violin plot
sns.violinplot(
    data=grouped,
    x='RedlineStatus',
    y='Score',
    hue='RedlineStatus',
    palette={'redline': 'red', 'not redline': 'lightblue'},
    legend=False
)

# Overlay individual data points
sns.stripplot(
    data=grouped,
    x='RedlineStatus',
    y='Score',
    hue='RedlineStatus',
    dodge=False,
    palette={'redline': 'black', 'not redline': 'black'},
    jitter=True,
    size=7,
    legend=False,
    alpha=0.8
)

# Annotate p-values at bottom right
plt.text(
    1.0, grouped['Score'].min() - 3,
    f"T-test p = {t_pval:.4f}\nMann-Whitney p = {u_pval:.4f}",
    ha='right',
    fontsize=10,
    style='italic'
)

# Labels
plt.title("Score Distributions: Redline vs Non-Redline Schools")
plt.ylabel("Score")
plt.xlabel("School Type")
plt.tight_layout()
plt.show()
```
